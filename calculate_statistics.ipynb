```python
import csv

# Function to read the CSV file and prepare data
def read_emissions_data(file_path):
    data = {}
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        headers = next(reader)  # Read the header row
        countries_column = headers[0]  # First column is 'CO2 per capita' (countries)
        years = headers[1:]  # Remaining columns are years

        for row in reader:
            country = row[0]
            for i, year in enumerate(years):
                if year not in data:
                    data[year] = []
                try:
                    data[year].append((country, float(row[i + 1])))  # Add country and emission
                except ValueError:
                    pass  # Skip invalid data
    return data

# Function to calculate statistics for a given year
def calculate_statistics(data, year):
    year = str(year)
    if year not in data:
        print(f"Data for the year {year} is not available.")
        return

    emissions = data[year]
    min_country, min_emission = min(emissions, key=lambda x: x[1])
    max_country, max_emission = max(emissions, key=lambda x: x[1])
    avg_emission = sum(x[1] for x in emissions) / len(emissions)

    print(f"Statistics for the year {year}:")
    print(f"  Minimum CO2 emission: {min_emission} (Country: {min_country})")
    print(f"  Maximum CO2 emission: {max_emission} (Country: {max_country})")
    print(f"  Average CO2 emission: {avg_emission:.2f}")

# Main Program
file_path = 'Emissions.csv'
emissions_data = read_emissions_data(file_path)
user_year = int(input("Enter a year (1997 to 2010): "))
calculate_statistics(emissions_data, user_year)
```
